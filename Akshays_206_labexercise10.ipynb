```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
import re

def validate_form():
    name = name_entry.get()
    email = email_entry.get()
    phone = phone_entry.get()
    gender = gender_var.get()
    year_of_birth = year_of_birth_spinbox.get()

    name_pattern = r'^[A-Za-z\s]+$'
    email_pattern = r'^\w+@[a-zA-Z_]+\.[a-zA-Z]{2,3}$'
    phone_pattern = r'^\d{10}$'

    if not re.match(name_pattern, name):
        messagebox.showerror("Validation Error", "Invalid Name")
        return

    if not re.match(email_pattern, email):
        messagebox.showerror("Validation Error", "Please Enter correct Email")
        return

    if not re.match(phone_pattern, phone):
        messagebox.showerror("Validation Error", "Phone Number has to have 10 numbers")
        return
    
    success_message = f"Form Submitted Successfully!\nName: {name}\nEmail: {email}\nPhone: {phone}\nGender: {gender}\nYear of Birth: {year_of_birth}"
    messagebox.showinfo("Success", success_message)

app = tk.Tk()
app.title("Sports Management Form")
app.geometry('400x400')

frame = ttk.Frame(app, padding=10)
frame.grid(row=0, column=0)

name_label = ttk.Label(frame, text="Player Name:")
name_entry = ttk.Entry(frame)

email_label = ttk.Label(frame, text="Player Email Id:")
email_entry = ttk.Entry(frame)

phone_label = ttk.Label(frame, text="Phone number:")
phone_entry = ttk.Entry(frame)

gender_label = ttk.Label(frame, text="Player Gender:")
gender_var = tk.StringVar()
gender_options = ["Male", "Female", "Other"]
gender_dropdown = ttk.Combobox(frame, textvariable=gender_var, values=gender_options)


sport_label = ttk.Label(frame, text="Favorite Sport:")
sport_var = tk.StringVar()
sport_options = ["Cricket", "FootBall", "BasketBall", "Hockey"]
sport_dropdown = ttk.Combobox(frame, textvariable=sport_var, values=sport_options)


year_of_birth_label = ttk.Label(frame, text="Year of Birth:")
year_of_birth_spinbox = ttk.Spinbox(frame, from_=1900, to=2023)

submit_button = ttk.Button(frame, text="Submit", command=validate_form)

# Place widgets using grid layout
name_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")
name_entry.grid(row=0, column=1, padx=5, pady=5)

email_label.grid(row=1, column=0, padx=5, pady=5, sticky="w")
email_entry.grid(row=1, column=1, padx=5, pady=5)

phone_label.grid(row=2, column=0, padx=5, pady=5, sticky="w")
phone_entry.grid(row=2, column=1, padx=5, pady=5)

gender_label.grid(row=3, column=0, padx=5, pady=5, sticky="w")
gender_dropdown.grid(row=3, column=1, padx=5, pady=5)

sport_label.grid(row=5, column=0, padx=5, pady=5, sticky="w")
sport_dropdown.grid(row=5, column=1, padx=5, pady=5)

year_of_birth_label.grid(row=6, column=0, padx=5, pady=5, sticky="w")
year_of_birth_spinbox.grid(row=6, column=1, padx=5, pady=5)

submit_button.grid(row=7, columnspan=2, padx=5, pady=10)

app.mainloop()
```
