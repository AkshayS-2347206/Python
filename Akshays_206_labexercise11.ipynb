```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('fifa_eda_stats.csv')
df.head(10)
```

```python
df.corr()
```

```python
corr = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr)
plt.show()
```

```python
x = df["Position"][0:20]
y = df["Name"][0:20]
plt.scatter(x,y)
plt.show()
```

```python
x = df["Name"][0:5]
y = df["Penalties"][0:5]
plt.xlabel('Name')
plt.ylabel('Penalties')
plt.title('Player Penalties')
plt.bar(x, y)
plt.show()
```

```python
plot = df["Skill Moves"].value_counts().plot(kind="bar")
plt.xlabel('Skills')
plt.ylabel('Count')
plt.title('Player Skill Sets')
plt.show()
```

```python
plot = df["Weight"].value_counts().plot(kind="line")
plt.ylabel("Weight")
plt.title('Player Weights')
plt.show()
```
